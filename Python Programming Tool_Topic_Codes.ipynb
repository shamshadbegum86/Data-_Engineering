```python
5. Python Nested Conditions
```

```python
marks = 72

if marks >= 50:
    print("Pass")
    if marks >= 90:
        print("Grade A+")
    elif marks >= 70:
        print("Grade A")
    else:
        print("Grade B")
else:
    print("Fail")
```

```python
8. OOPs Python
```

```python
class Car:
    def __init__(self, brand, year):
        self.brand = brand
        self.year = year
    
    def start(self):
        print(self.brand, "is starting...")

my_car = Car("Toyota", 2022)
my_car.start()
```

```python
class Student:
    def __init__(self, name, course):
        self.name = name
        self.course = course

    def show(self):
        print("Name:", self.name)
        print("Course:", self.course)

s1 = Student("John", "Python")
s1.show()
```
