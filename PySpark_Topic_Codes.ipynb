```python
1. PySpark Fundamentals
```

```python
# Import PySpark
from pyspark.sql import SparkSession

# Step 1: Create Spark Session
spark = SparkSession.builder \
    .appName("PySpark Fundamentals") \
    .master("local[*]") \
    .getOrCreate()

# Step 2: Sample data
data = [(1, "Alice"), (2, "Bob"), (3, "Charlie")]

# Step 3: Create DataFrame
df = spark.createDataFrame(data, ["id", "name"])

# Step 4: Show DataFrame
print("DataFrame content:")
df.show()

# Step 5: Display schema
print("DataFrame schema:")
df.printSchema()

# Step 6: Convert to Pandas safely
try:
    pandas_df = df.toPandas()
    print("Pandas DataFrame:")
    print(pandas_df)
except Exception as e:
    print("Error converting to Pandas:", e)
```

```python
10. Window Functions
```

```python
from pyspark.sql.window import Window
import pyspark.sql.functions as F

w = Window.orderBy("salary")

df = spark.createDataFrame([(1,"A",1000),(2,"B",1200)],["id","name","salary"])
df.withColumn("rank", F.rank().over(w)).show()
```

```python
13. Workflow Orchestration (Airflow Example)
```

```python
from airflow import DAG
from airflow.operators.python import PythonOperator

def etl_job():
    from pyspark.sql import SparkSession
    spark = SparkSession.builder.appName("AirflowETL").getOrCreate()
    df = spark.read.json("data.json")
    df.write.parquet("out/")
    
with DAG("spark_etl", schedule_interval="daily") as dag:
    task = PythonOperator(task_id="run_etl", python_callable=etl_job)
```

```python
19. UDFs and Performance Tuning
```

```python
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType

@udf(returnType=StringType())
def upper_case(x):
    return x.upper()

df.withColumn("upper_name", upper_case("name")).show()
```

```python
20. Data Quality and Validation
```

```python
import pyspark.sql.functions as F

# Check for nulls
df.select([F.count(F.when(F.col(c).isNull(), c)).alias(c) for c in df.columns]).show()

# Enforce schema
required_schema = ["id", "name"]
assert all(col in df.columns for col in required_schema)
```
