```python
3. AWS S3 Bucket
```

```python
import boto3

s3 = boto3.client('s3')

# Create bucket
s3.create_bucket(Bucket="my-demo-bucket-123")

# Upload file
s3.upload_file("test.txt", "my-demo-bucket-123", "test.txt")

# List objects
response = s3.list_objects_v2(Bucket="my-demo-bucket-123")
print(response)
```

```python
4. AWS Lambda – Invoke Lambda Function
```

```python
import boto3
import json

lambda_client = boto3.client('lambda')

response = lambda_client.invoke(
    FunctionName='myLambdaFunction',
    Payload=json.dumps({"message": "Hello Lambda"})
)

print(response['Payload'].read().decode())
```

```python
5. AWS Glue – Run ETL Job
```

```python
import boto3

glue = boto3.client('glue')

response = glue.start_job_run(
    JobName='my_glue_job'
)

print("Job Run ID:", response['JobRunId'])
```

```python
7. AWS Athena – Run SQL Query
```

```python
import boto3

athena = boto3.client('athena')

query = """
SELECT * FROM mydatabase.mytable LIMIT 10;
"""

response = athena.start_query_execution(
    QueryString=query,
    QueryExecutionContext={'Database': 'mydatabase'},
    ResultConfiguration={'OutputLocation': 's3://my-athena-output/'}
)

print("Execution ID:", response['QueryExecutionId'])
```

```python
8. AWS Kinesis – Send Data to Stream
```

```python
import boto3
import json

kinesis = boto3.client('kinesis')

data = {"event": "login", "user": "alice"}

kinesis.put_record(
    StreamName="MyStream",
    Data=json.dumps(data),
    PartitionKey="1"
)

print("Data sent to Kinesis")
```

```python
9. AWS Redshift – Run SQL
```

```python
import boto3

redshift = boto3.client("redshift-data")

response = redshift.execute_statement(
    ClusterIdentifier='redshift-cluster-1',
    Database='dev',
    Sql='SELECT * FROM users LIMIT 5;',
)

print("Statement ID:", response["Id"])
```

```python
10. AWS EventBridge – Create Rule
```

```python
import boto3

events = boto3.client("events")

response = events.put_rule(
    Name="MyScheduledRule",
    ScheduleExpression="rate(5 minutes)"
)

print("Rule ARN:", response["RuleArn"])
```

```python
11. AWS CloudWatch – Put Log Events
```

```python
import boto3
import time

logs = boto3.client('logs')

logs.put_log_events(
    logGroupName="my-log-group",
    logStreamName="my-stream",
    logEvents=[
        {
            "timestamp": int(time.time() * 1000),
            "message": "Hello from CloudWatch!"
        }
    ]
)
```

```python
12. AWS SageMaker – Start Training Job
```

```python
import boto3

sagemaker = boto3.client('sagemaker')

sagemaker.create_training_job(
    TrainingJobName='my-training-job',
    AlgorithmSpecification={
        'TrainingImage': '123456789012.dkr.ecr.us-east-1.amazonaws.com/myimage:latest',
        'TrainingInputMode': 'File'
    },
    RoleArn='arn:aws:iam::123456789012:role/SageMakerRole',
    InputDataConfig=[
        {
            'ChannelName': 'train',
            'DataSource': {
                'S3DataSource': {
                    'S3DataType': 'S3Prefix',
                    'S3Uri': 's3://mybucket/train/',
                    'S3DataDistributionType': 'FullyReplicated'
                }
            },
        }
    ],
    OutputDataConfig={
        'S3OutputPath': 's3://mybucket/output/'
    },
    ResourceConfig={
        'InstanceType': 'ml.m5.large',
        'InstanceCount': 1,
        'VolumeSizeInGB': 10
    }
)

print("Training job started.")
```
